```python
results_path = '/media/diskA/enric/parirset_models/results/'
df1 = pd.read_excel(pjoin(results_path, 'b1.xlsx'))
df2 = pd.read_excel(pjoin(results_path, 'd1.xlsx'))
df3 = pd.read_excel(pjoin(results_path, 'd2.xlsx'))
df4 = pd.read_excel(pjoin(results_path, 'd3.xlsx'))
df5 = pd.read_excel(pjoin(results_path, 'd4.xlsx'))
df6 = pd.read_excel(pjoin(results_path, 'e03_dry.xlsx'))
df7 = pd.read_excel(pjoin(results_path, 'cactest_full.xlsx'))
df8 = pd.read_excel(pjoin(results_path, 'cactest_none.xlsx'))

df = pd.concat((df6, df1, df4, df2, df3, df5, df6, df7, df8 ))
```

```python
df = df.replace('e03_dry', 'No reverb')
df = df.replace('b1', 'SE (Speech Enhancement)')
df = df.replace('d1', 'D1 (SE+beyer)')
df = df.replace('d2', 'D2 (D1+zylia_caps)')
df = df.replace('d3', 'D3 (D2+zylia_beams)')
df = df.replace('d4', 'D4 (D3+pa_permute)')

df.rename(columns={'sdr': 'SDR'}, inplace=True)
df.rename(columns={'model_name': 'Training RIRs'}, inplace=True)
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))

sns.pointplot(data=df, x="SDR", y="Training RIRs", capsize=0.2, errorbar=('ci', 95), palette="colorblind")
plt.grid(True)
plt.xlim([6, 9])
plt.tight_layout()

plt.savefig('boxplot_sdr.pdf', format='pdf', bbox_inches='tight')
```

```python
# 3. Generem el Pointplot horitzontal
# He posat 'Training RIRs' a l'eix Y i 'SDR' a l'eix X per seguir el teu estil
sns.pointplot(
    data=df, 
    x="SDR", 
    y="Training RIRs", 
    capsize=0.2,           # L'amplada de les barres horitzontals als extrems
    errorbar=('ci', 95),   # L'interval de confiança del 95%
    palette="colorblind",  # Paleta accessible per a daltonisme
    join=False,            # No connecta els punts amb una línia (més net per a categories)
    markers="D",           # Utilitza diamants en lloc de cercles (opcional, molt elegant)
    scale=1.2              # Fa els punts i línies una mica més gruixuts
)

# 4. Personalització addicional
plt.grid(True, axis='x', linestyle='--', alpha=0.7) # Grid només a l'eix X per guiar la vista
plt.title('Estimació de la Mitjana d\'SDR amb Interval de Confiança del 95%', pad=20)
plt.xlabel('SDR (dB)')
plt.ylabel('Training RIRs')

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))

# Invertim x i y per fer-lo horitzontal
sns.boxplot(
    data=df, 
    x='sdr',           # Dada numèrica a l'eix horitzontal
    y='model_name',    # Categories a l'eix vertical
    notch=True,        # Mantenim l'interval de confiança del 95%
    palette='Set2',
    showfliers=False,  # <--- Aquesta és la clau per treure els cercles
)

plt.title('Comparativa d\'SDR per Model (Format Horitzontal)')
plt.xlabel('SDR (dB)')
plt.ylabel('Model')
plt.tight_layout()
plt.savefig('boxplot_sdr.pdf')
plt.show()
```
