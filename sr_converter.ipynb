```python
import os
import soundfile as sf
import resampy
```

```python
target_sr = 44100 # we re
```

```python
files = os.listdir('/media/diskA/enric/parirset/test/')

for file in files:
    data, sr = sf.read('/media/diskA/enric/parirset/test/' + file)
    if sr != target_sr:
        data_resampled = resampy.resample(data.T, sr, target_sr)
    else:
        data_resampled = data.T
    data_resampled = data_resampled.T
    sf.write('/media/diskA/enric/parirset/test/' + file, data_resampled, target_sr)
```
